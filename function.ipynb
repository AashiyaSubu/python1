```python
def oddeven(num):
    if(num%2==0):
        print("Even")
        value="Even"
    else:
        print("Odd")
        value="ODD"
    return value
```

```python
callfn=oddeven(10)
print(callfn)
```

```python
def avrg (a,b,c,d,e):
    total=a+b+c+d+e
    percent=total/5
    return(percent)
    
```

```python
eng=int(input("Enter the English mark"))
tam=int(input("Enter the Tamil mark"))
mat=int(input("Enter the Maths mark"))
sci=int(input("Enter the Science mark"))
sst=int(input("Enter the SST mark"))

ans=avrg(eng,tam,mat,sci,sst)
print("your percentage is:")
print(ans)

```

```python
def triangle(a,b,c):
    area=(a*b)/2
    perimeter=a+b+c
    return(area,perimeter)
    
```

```python
side1=int(input("Enter the side1 mark"))
side2=int(input("Enter the side2 mark"))
side3=int(input("Enter the side3 mark"))

area,perimeter=triangle(side1,side2,side3)
print("Area of a triangle is")
print(area)
print("Perimeter of a triangle is")

print(perimeter)

```
