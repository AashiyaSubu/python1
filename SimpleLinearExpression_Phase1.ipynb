```python
# add librabry file pandas which is used to read files and tables
import pandas as pd
```

```python
#Read the datas and stored in avariable called dataset
dataset=pd.read_csv("Salary_Data.csv")
dataset
```

```python
#save each column in seperate variable so that we can split i/p and o/p

independent=dataset[["YearsExperience"]]
independent
```

```python
#Split the training set and test set using library file class name is sklearn using train_test_split fn
#it has four parameters and assigning these four parameters into four different variable
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=.30,random_state=0)
x_train
x_test
y_train
y_test
```

# Model creation using the fn Linear regression ,fit and calculate the weight(slope and bias(initial # value)

weight


```python
#Split the training set and test set using library file class name is sklearn using train_test_split fn
#it has four parameters and assigning these four parameters into four different variable
from sklearn.linear_model import LinearRegression
# regressor is universal variable name to load linear regression file
regressor= LinearRegression()
# fit fn is used calculate the weight(slope) and intial vale bias
regressor.fit(x_train,y_train)
weight=regressor.coef_
weight
```

```python
# bias is the minimum value means constant value of the slope this value is th o/p of 0 year experience salary
bias=regressor.intercept_
bias
```

```python
# Using this model creation evaluate the test values of i/p variable  and find the predicted values of the respected i/p variable
y_prediction=regressor.predict(x_test)
y_prediction
```

```python
# check predicted value with actual value and calculate R score.If R score value is approximately equal to 1 means, it is the beast mode
# using some lib file r2 score,metrics
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_prediction)
r_score
# this model find the value y=9360.26128619 * X + 26777.3913412
```
